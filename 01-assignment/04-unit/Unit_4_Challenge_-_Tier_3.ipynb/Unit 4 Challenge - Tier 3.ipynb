**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Writing a function that extracts average price for 1998 and 2018 and calculate the ratio between the years
def price_ratio(d):
    year_1998 = float(d['Average_price'][d['Year']==1998])
    year_2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [year_2018/year_1998]
    return ratio
    
```

```python
# Now calling the function to test a single observation. 
price_ratio(df1[df1['London_Borough']=='Barnet'])
```

```python
# Initializing the empty dictionary: new_dict.
new_dict = {}
```

```python
# Iterating over London Borough each observations.
for i in df1['London_Borough'].unique():
    # 
    borough = df1[df1['London_Borough'] == i]
    # 
    new_dict[i] = price_ratio(borough) 
print(new_dict)
```

```python
# Make a variable called df_ratios, and assign it the result of calling the DataFrame method on the dictionary final. 
df_ratios = pd.DataFrame(new_dict)
df_ratios
```

```python
# Flipping the dataframe and assigning it to df_ratios_t and reseting the index
df_ratios_t = df_ratios.T
df_ratios = df_ratios_t.reset_index()
df_ratios.head()
```

```python
# Renaming the index and 0 column
df_ratios.rename(columns={'index':'London_Borough', 0:'Ratio'}, inplace=True)
df_ratios.head()
```

```python
# Finally, sorting the variable 2018 which contains ratios in descending order to identify the most the borough that has greatest increase within these two decades.
highest10 = df_ratios.sort_values(by='Ratio',ascending=False).head(10)
```

```python
print(highest10)
```

```python
# Visualizing the result in the form of bar chart and assigning x axis with Borough
final = highest10[['London_Borough', 'Ratio']].plot(kind='bar', color='orange')

final.set_xticklabels(highest10.London_Borough)
```
